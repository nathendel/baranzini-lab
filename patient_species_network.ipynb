```python
useful_cols = ['SampleID'] + ['Disease_state'] + bact_cols
useful_df = df[useful_cols]
#g=useful_df.groupby("Disease_state")
#g.head()
ctrl_df = useful_df[useful_df.Disease_state == 'Control']
ms_df = useful_df[useful_df.Disease_state == 'MS']
```

```python
ppl_combos_ctrl = list(it.combinations(ctrl_df.index,2))
ppl_combos_ms = list(it.combinations(ms_df.index,2))
```

```python
def shared_species_high(df, ppl_combos):
    rows = []
    for ppl_combo in ppl_combos:
        p1 = ppl_combo[0]
        p2 = ppl_combo[1]
        threshold = 0.085
        a = (df.ix[p1, 2:] > threshold) & (df.ix[p1, 2:] != 0)
        b = (df.ix[p2, 2:] > threshold) & (df.ix[p2, 2:] != 0)
        shared = sum(a & b)
        rows.append((p1, p2, shared))
    return rows

rows_ctrl_high = shared_species_high(ctrl_df,ppl_combos_ctrl)
rows_ms_high = shared_species_high(ms_df,ppl_combos_ms)

ctrl_inter_high_df=pd.DataFrame(rows_ctrl_high, columns = ['person1', 'person2', 'num_shared'])
ms_inter_high_df=pd.DataFrame(rows_ms_high, columns = ['person1', 'person2', 'num_shared'])
```

```python
def shared_species_low(df, ppl_combos):
    rows = []
    for ppl_combo in ppl_combos:
        p1 = ppl_combo[0]
        p2 = ppl_combo[1]
        threshold = 0.0002
        a = (df.ix[p1, 2:] < threshold) & (df.ix[p1, 2:] != 0)
        b = (df.ix[p2, 2:] < threshold) & (df.ix[p2, 2:] != 0)
        shared = sum(a & b)
        rows.append((p1, p2, shared))
    return rows

rows_ctrl_low = shared_species_low(ctrl_df,ppl_combos_ctrl)
rows_ms_low = shared_species_low(ms_df,ppl_combos_ms)

ctrl_inter_low_df=pd.DataFrame(rows_ctrl_low, columns = ['person1', 'person2', 'num_shared'])
ms_inter_low_df=pd.DataFrame(rows_ms_low, columns = ['person1', 'person2', 'num_shared'])
```

```python
ctrl_inter_low_df = ctrl_inter_low_df[ctrl_inter_low_df.num_shared != 0]
ctrl_inter_high_df = ctrl_inter_high_df[ctrl_inter_high_df.num_shared != 0]

ms_inter_low_df = ms_inter_low_df[ms_inter_low_df.num_shared != 0]
ms_inter_high_df = ms_inter_high_df[ms_inter_high_df.num_shared != 0]
```
